```python
import requests
import urllib.request as ureq
from bs4 import BeautifulSoup
```

```python
url = 'https://yts.lt/browse-movies/0/all/all/0/year'
response = requests.get(url)
response
```

```python
res=BeautifulSoup(response.text,"html.parser")
```

```python
res
```

```python
print(res.prettify())
```

```python
photos= res.findAll('div','browse-movie-wrap col-xs-10 col-sm-4 col-md-5 col-lg-4')
```

```python
photos
```

```python
arr=input("Enter The name = ")
```

```python
arr="anya"
for i in photos:
    a=i.find('figure')
    url_img=a.img['src']
    name=url_img.split('/')
    name=name[-2]+"_"+name[-1]
    ureq.urlretrieve(url_img,name)
    if(arr in name):
        print(name)
        print(url_img)
```
